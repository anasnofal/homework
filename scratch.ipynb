```python
import pandas as pd
import os 
```

```python
data=pd.DataFrame()
parse_dates = ['lpep_pickup_datetime', 'lpep_dropoff_datetime']

for month in [10,11,12]:
    url = f'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/green/green_tripdata_2020-{month}.csv.gz'
    csv_name= f'green_taxi_{month}.csv.gz'
    os.system(f"wget {url} -O {csv_name}")
    
    df=pd.read_csv(csv_name,sep=',',compression="gzip",parse_dates=parse_dates)
    data= pd.concat([data,df],ignore_index=True)
```

```python
data.info()
```

```python
data['lpep_pickup_date']= data['lpep_pickup_datetime'].dt.date
data.columns = (data.columns
                .str.replace('(?<=[a-z])(?=[A-Z])', '_', regex=True)
                .str.lower()
             )
data=data[data['passenger_count']>0]
data=data[data['trip_distance']>0]
data
```

```python
assert 'vendor_id' in data.columns, 'vendor_id does not exist'
assert data['trip_distance'][data['trip_distance']==0].sum()==0, "there are zeros"
assert data['passenger_count'][data['passenger_count']==0].sum()==0, "there are zeros"
```
